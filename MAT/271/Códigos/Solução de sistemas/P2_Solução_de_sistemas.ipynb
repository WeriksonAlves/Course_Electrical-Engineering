```python
def cria_matrizes(tamanho):
    
    matriz_A = np.zeros((tamanho,tamanho)) #[] # lista vazia
    matriz_b = np.zeros((tamanho,1)) #[] # lista vazia
    
    #Matriz A
    for i in range(tamanho):
        for j in range(tamanho):
            matriz_A[i][j] = input('Entre com o elemento da matriz A_ij: ')
    
    #Matriz b
    for i in range(tamanho):
        for j in range(1):
            matriz_b[i][j] = input('Entre com o elemento do vetor b_i0: ')
        
    return matriz_A, matriz_b
#...............................................................................................................................
def formata_matriz(M):
    m = len(M) # número de linhas
    n = len(M[0]) # número de colunas
    s = ""
    for i in range(m):
        for j in range(n):
           s += "%9.4f " % M[i][j]
        s += "\n"
    return s
#...............................................................................................................................
def parada(x,x1,tamanho,tipo):
    dif = []
    dif1 = []
    for i in range(tamanho):
        diferenca = abs(x1[0][i] - x[0][i])
        diferenca1 = abs(x1[0][i])
        dif.append(diferenca)
        dif1.append(diferenca1)
    if (tipo == 0):#absoluto
        erro = max(dif)
    elif (tipo == 1):#relativo
        erro = max(dif)/max(dif1)
    return erro
#...............................................................................................................................
```

```python
def fatoraLU(A):  
    U = np.copy(A)  
    n = np.shape(U)[0]  
    L = np.eye(n)  
    for j in np.arange(n-1):  
        for i in np.arange(j+1,n):  
            L[i,j] = U[i,j]/U[j,j]  
            for k in np.arange(j+1,n):  
                U[i,k] = U[i,k] - L[i,j]*U[j,k]  
            U[i,j] = 0  
    return L, U
#...............................................................................................................................
def substituicoes_sucessivas(A, b):
    '''Executa o método das substituições retroativas para resolver o sistema Ax=b.
       Parâmetros de entrada: A é uma matriz triangular superior e b é o vetor constante.
    '''
    ## n é a ordem da matriz A
    n = len(A)
    
    ## inicializa o vetor x com tamanho n e elementos iguais a 0
    x = n * [0] 
    
    for i in range (0,n):
        S=0
        for j in range(0,i):
            S = S + A[i][j] * x[j]
        x[i] = (b[i] - S)/A[i][i]       
    
    return x
#...............................................................................................................................
def substituicoes_retroativas(A, b):
    '''Executa o método das substituições retroativas para resolver o sistema Ax=b.
       Parâmetros de entrada: A é uma matriz triangular superior e b é o vetor constante.
    '''
    ## n é a ordem da matriz A
    n = len(A)
    
    ## inicializa o vetor x com tamanho n e elementos iguais a 0
    x = n * [0] 
    
    for i in range (n-1,-1,-1):
        S=0
        for j in range(i+1,n):
            S = S + A[i][j] * x[j]
        x[i] = (b[i] - S)/A[i][i]       
    
    return x
#...............................................................................................................................
def lux(L,U,b):
    '''
    Resolve o sistema LUx=b.
    Esse método resolve os dois sistemas lineares triangulares.
    '''
    
    y = substituicoes_sucessivas(L,b)
    
    x = substituicoes_retroativas(U,y)
    
    return x, y
#...............................................................................................................................
def verifica_lu(A):
    n = len(A)
    for v in range(1,n+1):
        delta_k = np.zeros((v,v))
        for i in range (0,v):
            for j in range(0,v):
                delta_k[i][j] = A[i][j]

        det = np.linalg.det(delta_k)
        if (det == 0):
            return 0
        else:
            print('\n Delta',v,': \n%s' % formata_matriz(delta_k)) 
            print('det(Delta',v,') = ',det)
#...............................................................................................................................
def metodo_LU(tamanho):
    print('..........Método da decomposição LU..........')
    print('\n Entrando com as matrizes A e b:')
    A_LU, b_LU = cria_matrizes(tamanho)
    
    print('\n Após as definições temos as seguintes matrizes:')
    print("A: \n%s" % formata_matriz(A_LU))
    print("b: \n%s" % formata_matriz(b_LU))
    
    print('Primeiramente devemos verificar se é possivel aplicar este método, logo, aplicando os "Menores Principais", temos:')
    
    verifica = verifica_lu(A_LU)
    if (verifica == 0):
        print('Não é possivel aplicar este método.')
    else:
        print('\n Com todos os Deltas diferentes de zeros podemos prosseguir com o método.')
        
        L, U = fatoraLU(A_LU)
        print('\n Após a decomposição LU obtemos as seguintes matrizes:')
        print("L: \n%s" % formata_matriz(L))
        print("U: \n%s" % formata_matriz(U))

        x, y = lux(L,U,b_LU)
        print('\n Após os cálculos encontramos os vetores y e x:')
        print("y: \n%s" % formata_matriz(y))
        print("x: \n%s" % formata_matriz(x))
```

```python
def calcula_JR(tamanho,A,y,x,tipo):
    x1 = np.zeros((1,tamanho))
    for i in range(tamanho):
        S = 0
        for j in range(tamanho):
            if ( i == j):
                S += y[i][0]/A[i][i]
            elif (i != j):
                S -= (A[i][j]/A[i][i]) * x[0][j]
        x1[0][i] = S
    erro = parada(x,x1,tamanho,tipo)
    
    return x1, erro
#...............................................................................................................................
def convergencia_JR(tamanho,A):
    beta = np.zeros((tamanho,1))
    for i in range(tamanho):
        B = 0
        for j in range(tamanho):
            if (i != j):
                B = B + abs(A[i][j]/A[i][i])            
        beta[i][0] = B
    
    print("\nBeta: \n%s" % formata_matriz(beta))
    for a in range (tamanho):  
        if (beta[a][0] < 1):
            return 1
        else:
            return 0
#...............................................................................................................................   
def metodo_JR(tamanho,tipo,E):
    print('..........MÉTODO DE JACOBI-RICHARDSON..........')
    print('\n Entrando com as matrizes A e y:')
    A, y = cria_matrizes(tamanho)
    
    print('\n Após as definições temos as seguintes matrizes:')
    print("A: \n%s" % formata_matriz(A))
    print("y: \n%s" % formata_matriz(y))
    
    if (tipo == 0):
        print('O erro escolhido foi o ERRO ABSOLUTO.')
    elif (tipo == 1):
        print('O erro escolhido foi o ERRO RELATIVO.')
    else:
        print('Campo preenchido incorretamente!')
        return None
    
    conv = convergencia_JR(tamanho,A)  
    if (conv == 0):
        print('Não há garantia de convergência (pelo critério "norma linha") para este sistema.')
    else:
        print('Garantida a convergência (pelo critério "norma linha") deste sistema, podemos seguir com o método.')
        k = 0
        x = [[1],[0],[1]]
        erro =1
        while (erro > E and k < 15 ):
            print('\n O vetor x_k é %\ns' % formata_matriz(x))
            x, erro = calcula_JR(tamanho,A,y,x,tipo)
            print('O erro é de %9.5f' % erro)
            k += 1
```
